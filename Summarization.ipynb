````python
# libraries
from transformers.utils import logging
logging.set_verbosity_error()
from transformers import pipeline
import torch
````

### Build the `summarization` pipeline using ðŸ¤— Transformers Library

````python
summarizer = pipeline(task="summarization",
                      model="facebook/bart-large-cnn",
                      torch_dtype=torch.bfloat16)
````

**Bfloat16** is a numerical data type that represents numbers using 16 bits (half the precision of a standard single-precision floating-point number, which uses 32 bits).
It offers a balance between storage efficiency and computational performance compared to full-precision (32-bit) floating-point numbers.

* Benefits for LLMs*

Reduced Memory Footprint: Since bfloat16 numbers take up half the space, they can significantly reduce the memory required to store the massive parameters and activations within an LLM. This is crucial for training and deploying large models on resource-constrained hardware.
Potential Speedup: In some cases, bfloat16 operations can achieve faster execution on hardware that supports them with specialized instructions. This can lead to quicker training and inference times for LLMs.

Model info: ['bart-large-cnn'](https://huggingface.co/facebook/bart-large-cnn)

````python
text = """The rivalry between the Pakistan Cricket Board (PCB) and the Board of Control for Cricket in India (BCCI)
 burns with an intensity that transcends the sport itself. Often referred to as the "Battleground,"
  matches between these two neighboring nations become a passionate display of national pride,
  fueled by the historical and political complexities of the region. Every encounter is a highly anticipated spectacle,
   watched by millions with hearts pounding and emotions running high.
   On the field, the competition is fierce and uncompromising.
    Both teams boast a rich cricketing heritage,
    with world-class players who have etched their names in cricketing folklore.
    Pakistan currently leads the dance in One Day Internationals (ODIs),
     having emerged victorious in a remarkable 73 matches compared to India's 57 wins.
      This dominance is particularly intriguing given India's overall success in ICC tournaments.
       However, the recent past has witnessed a shift in momentum.
        India has displayed a resurgent form in ODIs against Pakistan,
        culminating in a convincing win at the 2023 Cricket World Cup.
          With both teams constantly evolving and brimming with young talent,
          the future of this enthralling cricketing rivalry promises many more chapters filled with nail-biting finishes,
          heart-stopping moments, and passionate displays of fanaticism,
           making it one of the most captivating rivalries in the cricketing world."""
````

````python
summary = summarizer(text,
                     min_length=101,
                     max_length=150)
````
